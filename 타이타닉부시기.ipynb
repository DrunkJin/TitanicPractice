### 객실등급별 생존률 확인하기
#### 1 : 생존, 0 : 사망
#### Survived 평균이 높을수록 생존률이 높다고 할 수 있다.

```python
print("=============등급별 객실 갯수=============")
print(tr.groupby(by='Pclass').size(),'\n')
print("=============등급별 생존자&사망자 수=============")
print(tr.groupby(['Pclass','Survived']).size(),'\n')
print("=============객실 등급별 생존 평균=============")
print(tr.groupby(by='Pclass')['Survived'].mean(),'\n')
```

#### null값이 400개 이상들어가있는 행 제거 (절반이상 null이라면 의미없다고 판단)

```python
tr.dropna(axis=1, thresh=400, inplace=True)
tr
```

```python
print("Age null갯수 : ",tr['Age'].isnull().sum(),"\n")
print("Embarked null갯수 : ", tr["Embarked"].isnull().sum(),'\n')
```

#### Age의 빈값을 중앙값으로 채워넣기

```python
median_age = tr['Age'].median(axis=0)
print(median_age)
tr['Age'].fillna(median_age, inplace=True)
print("Age null갯수 : ",tr['Age'].isnull().sum())
```

```python
tr_train = tr.drop(['PassengerId',"Survived",'Name','Ticket','Fare'], axis=1)
tr_label = tr['Survived']
```

```python
tr_train.head()
```

### String Data를 Int Data로 바꾸기

```python
sex_dum = pd.get_dummies(tr['Sex'])
sex_dum = sex_dum.add_prefix('Sex_')
sex_dum.head()
```

```python
Emb_dum = pd.get_dummies(tr['Embarked'])
Emb_dum = Emb_dum.add_prefix('Embarked_')
Emb_dum.head()
```

#### tr_train에 변환한 데이터 붙이기

```python
tr_train.drop(['Sex','Embarked'],axis=1, inplace=True)
tr_train = pd.concat([tr_train, sex_dum, Emb_dum], axis=1)
tr_train.head()
```

#### Test 동일하게 전처리

```python
ts.drop(['Cabin'],axis=1,inplace=True)
median_age = ts['Age'].median(axis=0)
print(median_age)
ts['Age'].fillna(median_age, inplace=True)
print("Age null갯수 : ",ts['Age'].isnull().sum())

ts = ts.drop(['PassengerId','Name','Ticket','Fare'], axis=1)
sex_dum2 = pd.get_dummies(ts['Sex'])
sex_dum2 = sex_dum2.add_prefix('Sex_')
Emb_dum2 = pd.get_dummies(ts['Embarked'])
Emb_dum2 = Emb_dum2.add_prefix('Embarked_')
ts.drop(['Sex','Embarked'],axis=1, inplace=True)
ts = pd.concat([ts, sex_dum2, Emb_dum2], axis=1)
ts
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier(n_estimators=100, random_state=777)
edu_model = rfc.fit(tr_train, tr_label)
cross_val_score(rfc, tr_train, tr_label, cv=10).mean()
```

```python
ans = pd.DataFrame(edu_model.predict(ts))
```

```python
ans.columns=['Survived']
```

```python
paper = pd.read_csv("submission.csv")
```

```python
paper.drop(['Survived'],axis=1,inplace=True)
paper2 = pd.concat([paper,ans],axis=1)
paper2.to_csv("배운방법.csv", index=False)
```
